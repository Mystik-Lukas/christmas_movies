```python
import pandas as pd 
import altair as alt

f = "christmas_movies.csv"
data = pd.read_csv(f)

data.head()
```

```python
data_pie = data.sort_values(by=['imdb_rating'], ascending=False)

top_10_pie = data_pie.iloc[:10]

base_pie = alt.Chart(top_10_pie).encode(
    theta = alt.Theta(field="imdb_rating", type="quantitative", title="IMDB Rating", stack=True),
    color = alt.Color(field="title", type="nominal", title="Movie Title"),
    radius = alt.Radius("imdb_rating", scale=alt.Scale(type="sqrt", zero=True)),
).properties(
    title="Top 10 Movies of All Time By IMDB Rating",
    width=500,
    height=500
)

pie = base_pie.mark_arc(stroke="#fff", innerRadius=20)

textpie = base_pie.mark_text(radiusOffset=10).encode(
    text="imdb_rating"
)

pie + textpie 
```

```python
data_bar = data
df = data_bar.copy(True)

df['gross'] = df['gross'].apply(lambda x: x.replace(".", "").replace('$', '').replace("M", "0000") if isinstance(x, str) else x).astype(float)

df = df.sort_values(by=['gross'], ascending=False)

top_10_bar = df.iloc[:10]

base_bar = alt.Chart(top_10_bar).mark_bar().encode(
    y = alt.Y(field="gross", type="quantitative", title="Gross Revenue ($)", stack=True),
    x = alt.X(field="title", type="nominal", title="Movie Title"),
).properties(
    title="Top 10 Movies of All Time By Gross Revenue",
    width=500,
    height=500
)

textbar = base_bar.mark_text(baseline='middle', fontSize=7, dy=-10).encode(
    text=alt.Text("gross")
)

base_bar + textbar 
```
